```python
models = []
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('DTC', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
```

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score, GridSearchCV
```

```python
num_folds = 10
seed = 7
scoring = 'accuracy'
```

```python
results = []
names = []
for name, model in models:
    kfold = KFold(n_splits=num_folds, random_state=seed)
    cv_results = cross_val_score(model, xtrain, ytrain, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

```python
#We can see that Logistic Regression gives the highest accuracy score= 90%.
#KNN gives a score of 87%
#SVm gives a score of 85.5%
#LDA gives a score of 85.1%
```
